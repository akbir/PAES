```python
import pandas as pd
import os

files = os.listdir("mturk")

dfs = []
for file in files:
    if file.startswith("Batch"):
        dfs.append(pd.read_csv(f"mturk/{file}"))
        
df = pd.concat(dfs)
display(df)
```

```python
df = df.drop(
    columns=[
        "HITId",
        "HITTypeId",
        "AssignmentId",
        "Keywords",
        "Reward",
        "CreationTime",
        "MaxAssignments",
        "RequesterAnnotation",
        "AssignmentDurationInSeconds",
        "NumberOfSimilarHITs",
        "LifetimeInSeconds",
        "Title",
        "Description",
        "AssignmentId",
        "AssignmentStatus",
        "SubmitTime",
        "AutoApprovalTime",
        "ApprovalTime",
        "RejectionTime",
        "AutoApprovalDelayInSeconds",
        "AcceptTime",
        "WorkTimeInSeconds",
        "Last30DaysApprovalRate",
        "LifetimeApprovalRate",
        "Last7DaysApprovalRate",
        "Input.audio_url",
        "RequesterFeedback",
        "Expiration",
        "Approve",
        "Reject"
    ])

df


```

```python
# get annotation columns 

print(list(df.columns))


annotations = list(df.columns)
annotations = [anno for anno in df.columns if anno.startswith("Answer.")]
annotations = [anno.split(".")[1] for anno in annotations]
```

```python
# Mean Agreement over 

import simpledorff
import numpy as np

agreements = []
for annotation in annotations:
    
    agreement = simpledorff.calculate_krippendorffs_alpha_for_df(
        df,
        experiment_col='Input.audio_title',
        annotator_col='WorkerId',
        class_col=f'Answer.{annotation}'
    )
    
    agreements.append(agreement)
    print(f"{annotation}: {agreement}")

    
print(f"Mean Agreement: {np.mean(agreements)}")
```

```python
# remove ordinal component, single emotion

df2 = df.copy()

for annotation in annotations:
    df2[f"Answer.{annotation}"] = df2[f"Answer.{annotation}"].apply(lambda x: 1 if x>=1 else 0)
    
    
df2
```

```python
agreements = []
for annotation in annotations:
    agreement = simpledorff.calculate_krippendorffs_alpha_for_df(
        df2,
        experiment_col='Input.audio_title',
        annotator_col='WorkerId',
        class_col=f'Answer.{annotation}')
    print(f"{annotation}: {agreement}")

    agreements.append(np.abs(agreement))
    
print(f"Mean Agreement: {np.max(agreements)}")
```

```python
# remove ordinal component, multi emotion agreement

df3 = df.copy()

for annotation in annotations:
    df3[f"Answer.{annotation}"] = df3[f"Answer.{annotation}"].apply(lambda x: 1 if x>=1 else 0)
    
    
df3['Annotation'] = [np.zeros(len(annotations)) for i in range(len(df3))]

for row in df3.iterrows():
    row = row[1]
    for i, annotation in enumerate(annotations):
        row["Annotation"][i] = 1 if row[f"Answer.{annotation}"] == 1 else 0
        
df3['Annotation'] = df3['Annotation'].apply(lambda x: tuple(x))
```

```python
agreement = simpledorff.calculate_krippendorffs_alpha_for_df(
        df3,
        experiment_col='Input.audio_title',
        annotator_col='WorkerId',
        class_col=f'Annotation')

print(agreement)
```

```python
df4 = df.copy()    
df4['Annotation'] = [np.zeros(len(annotations)) for i in range(len(df3))]

for row in df4.iterrows():
    row = row[1]
    for i, annotation in enumerate(annotations):
        row["Annotation"][i] = 1 if row[f"Answer.{annotation}"] == 1 else 0
        
df4['Annotation'] = df4['Annotation'].apply(lambda x: tuple(x))
```

```python
agreement = simpledorff.calculate_krippendorffs_alpha_for_df(
        df4,
        experiment_col='Input.audio_title',
        annotator_col='WorkerId',
        class_col=f'Annotation')

print(agreement)
```

```python
### Pearson Correlation For Intensity
import itertools
import math

poems = df['Input.audio_title'].unique()
p_correlations = []

for poem in poems:
    poem_df = df[df['Input.audio_title']==poem].drop(
        columns=['Input.Author', 'Input.audio_title', "WorkerId"])
    
    annotators_per_poem_df= [s for i, s in poem_df.iterrows()]
    
    all_annotator_pairs = list(itertools.combinations(annotators_per_poem_df, 2))
    
    for s1, s2 in all_annotator_pairs:
        p_corr = np.abs(s1.corr(s2, method='pearson'))
        
        if math.isnan(p_corr): # catches empty rows
            print(s1, s2)
        else:
            p_correlations.append(p_corr)    
```

```python
print(np.mean(p_correlations))
```
